```python
import matplotlib.pyplot as plt

X = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'June', 'July']
Y = [0, 10, 15, 17, 23, 26, 30]

plt.plot(X, Y, "-.sb")
plt.xlabel('month')
plt.ylabel('temperature')
plt.title('Temperatures of Inchoen')
plt.show()
```
